```python
#for the above dataset, I need to apply Kmeans, aggloromative clustering and DBSCAN
#I will use the elbow method to find the optimal number of clusters
#I will use the silhouette score to find the optimal number of clusters
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
```

```python
def plot_data(df):
    plt.scatter(df['x1'], df['x2'])
    plt.xlabel('x1')  # replace 'x1' with your x1-axis label
    plt.ylabel('x2')  # replace 'x2' with your x2-axis label
    plt.title('Scatter plot of x1 vs x2')  # replace with your title
    plt.show()
```

```python
def plot_clusters(df,k):
    kmeans = KMeans(n_clusters=k, random_state=0).fit(df)
    df['cluster'] = kmeans.labels_
    plt.scatter(df['x1'], df['x2'], c=df['cluster'])
    plt.xlabel('x1')  # replace 'x1' with your x1-axis label
    plt.ylabel('x2')  # replace 'x2' with your x2-axis label
    plt.title('Scatter plot of x1 vs x2')  # replace with your title
    plt.show()
```

```python
plot_clusters(df,5)
```

```python
range_clusters = range(1, 11)
inertias=[]
for i in range_clusters:
    kmeans = KMeans(n_clusters=i, random_state=0).fit(df[['x1', 'x2']])
    inertias.append(kmeans.inertia_)

plt.plot(range_clusters, inertias, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
def plot_silhouette(df, range_clusters):
    sil = []
    for i in range_clusters:
        kmeans = KMeans(n_clusters=i, random_state=0).fit(df[['x1', 'x2']])
        labels = kmeans.labels_
        sil.append(silhouette_score(df[['x1', 'x2']], labels, metric='euclidean'))
    plt.plot(range_clusters, sil, 'bx-')
    plt.xlabel('Number of clusters')
    plt.ylabel('Silhouette Score')
    plt.title('The Silhouette Method showing the optimal k')
    plt.show()
```

### From the above plot optimum number of clusters is 5

```python
#now silhoute score
plot_silhouette(df, range(2, 11))
```

```python
def print_silhouette(df, k):
    #silhouette_score for k=5
    kmeans = KMeans(n_clusters=k, random_state=0).fit(df[['x1', 'x2']])
    silhouette_score(df[['x1', 'x2']], kmeans.labels_)
    print(f"silhouette_score for K={k} is",silhouette_score(df[['x1', 'x2']], kmeans.labels_))
```

```python
print_silhouette(df, 5)
```

### Aggloromative Clustering

```python
#now Agglomerative clustering
df = pd.read_csv('./Dataset/Dataset2/Compact.csv')
```

```python
def print_silhouette_agg(df, k):
    #silhouette_score for k=5
    agg_clustering = AgglomerativeClustering(n_clusters=k).fit(df[['x1', 'x2']])
    silhouette_score(df[['x1', 'x2']], agg_clustering.labels_)
    print(f"silhouette_score for K={k} is",silhouette_score(df[['x1', 'x2']], agg_clustering.labels_))
```

```python
def plot_agglomerative_clusters(df,k):
    agg_clustering = AgglomerativeClustering(n_clusters=k).fit(df[['x1', 'x2']])

    df['cluster'] = agg_clustering.labels_

    plt.scatter(df['x1'], df['x2'], c=df['cluster'])
    plt.show()
```

```python
plot_agglomerative_clusters(df,5)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
def plot_dendrogram(df):
    linked = linkage(df[['x1', 'x2']], 'single')

    plt.figure(figsize=(10, 7))
    dendrogram(linked,
               orientation='top',
               distance_sort='descending',
               show_leaf_counts=True)
    plt.show()
```

```python
plot_dendrogram(df)
```

```python
def plot_silhouette_agg(df, range_clusters):
    sil = []
    for i in range_clusters:
        agg_clustering = AgglomerativeClustering(n_clusters=i).fit(df[['x1', 'x2']])
        labels = agg_clustering.labels_
        sil.append(silhouette_score(df[['x1', 'x2']], labels, metric='euclidean'))
    plt.plot(range_clusters, sil, 'bx-')
    plt.xlabel('Number of clusters')
    plt.ylabel('Silhouette Score')
    plt.title('The Silhouette Method showing the optimal k')
    plt.show()
```

```python
plot_silhouette_agg(df, range(2, 11))
```

### From the above plot optimal number of clusters is 5

```python
print_silhouette_agg(df, 5)
```

### DBScan

#### Compact.csv

```python
#now DBSCAN
df = pd.read_csv('./Dataset/Dataset2/Compact.csv')

def plot_dbscan_clusters(df, eps, min_samples):
    dbscan = DBSCAN(eps=eps, min_samples=min_samples).fit(df[['x1', 'x2']])
    df['cluster'] = dbscan.labels_
    plt.scatter(df['x1'], df['x2'], c=df['cluster'])
    plt.show()
def print_noise_points(df,eps,min_samples):
    dbscan = DBSCAN(eps=eps, min_samples=min_samples).fit(df[['x1', 'x2']])
    labels = dbscan.labels_
    print(f"Number of noise points for eps={eps} and min_samples={min_samples} is",list(labels).count(-1))

def find_optimal_eps_min_samples(df):
    eps_values = range(1, 21)
    min_samples_values = range(1, 101)
    sil = []
    for eps in eps_values:
        for min_samples in min_samples_values:
            dbscan = DBSCAN(eps=eps, min_samples=min_samples).fit(df[['x1', 'x2']])
            labels = dbscan.labels_
            # Check if there are at least 2 clusters
            if len(set(labels)) > 1:
                sil.append((silhouette_score(df[['x1', 'x2']], labels, metric='euclidean'), eps, min_samples))
            else:
                sil.append((float('-inf'), eps, min_samples))

    sil.sort(reverse=True)
    eps_optimal = sil[0][1]
    min_samples_optimal = sil[0][2]
    print(f"Optimal eps is {eps_optimal} and optimal min_samples is {min_samples_optimal}")
    return eps_optimal, min_samples_optimal

plot_dbscan_clusters(df, 4, 20)
```

```python

def print_silhouette_dbscan(df, eps, min_samples):
    dbscan = DBSCAN(eps=eps, min_samples=min_samples).fit(df[['x1', 'x2']])
    labels = dbscan.labels_
    sil = silhouette_score(df[['x1', 'x2']], labels, metric='euclidean')
    print(f"silhouette_score for eps={eps} and min_samples={min_samples} is",sil)

print_silhouette_dbscan(df, 4, 20)
```

```python
print_noise_points(df, 4, 20)
```

```python
from mpl_toolkits.mplot3d import Axes3D
eps_values = range(1, 6)
min_samples_values = range(1, 51)
sil = []
for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples).fit(df[['x1', 'x2']])
        labels = dbscan.labels_
        # Check if there are at least 2 clusters
        if len(set(labels)) > 1:
            sil.append(silhouette_score(df[['x1', 'x2']], labels, metric='euclidean'))
        else:
            sil.append(float('-inf'))
```

```python
eps_values, min_samples_values = np.meshgrid(eps_values, min_samples_values)

# Convert the silhouette scores to a 2D array
silhouette_scores = np.array(sil).reshape(eps_values.shape)

# Create a 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(eps_values, min_samples_values, silhouette_scores)

ax.set_xlabel('eps')
ax.set_ylabel('min_samples')
ax.set_zlabel('Silhouette Score')

plt.show()
```

#### For the compact dataset, all the algorithms performed reasonably well. and overall, Kmeans performed well on all datasets. In skewed.csv dataset, agglomerative kmeans and dbscan performed relatively better than agglomerative. Agglomerative clustering can struggle with skewed datasets due the way it forms clusters. In skewed dataset, data points are not uniformly distributed and may have a long tail on one side. This can cause agglomerative clustering to create clusters that are not meaningful or representative of the underlying data distribution.
