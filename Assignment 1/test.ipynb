```python
fifa = pd.read_csv('fifa.csv')
```

```python
fifa.head(5)
```

```python
fifa.info()
```

```python
fifa.columns
```

```python
bins = [x for x in range(40, 101, 10)]
plt.hist(fifa.Overall,bins=bins)
plt.xticks(bins)
plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.show()
```

```python
fifa['Preferred Foot'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(8,8))    
```

```python
#another method
left_foot = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right_foot = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
plt.figure(figsize=(8,5))
plt.pie([left_foot, right_foot], labels=['Left', 'Right'], autopct='%.2f %%', colors=['#abcdef', '#aabbcc'])
```

```python
def fun(x):
    if type(x)==str:
        return int(x.replace('lbs',''))
    return x
fifa['Weight'] = fifa['Weight'].apply(fun)
```

```python
bins=[x for x in range(90,251,10)]
plt.hist(fifa.Weight, bins=bins, color='#abcdef')
plt.xticks(bins)
plt.ylabel('Number of Players')
plt.xlabel('Weight (lbs)')
plt.title('Weight Distribution of FIFA Players (in lbs)')
plt.show()
```

```python
light_weight = len(fifa.loc[fifa['Weight'] < 125 ] )
medium_weight = len(fifa.loc[(fifa['Weight'] >= 125) & (fifa['Weight'] < 200)])
heavy_weight = len(fifa.loc[fifa['Weight'] >= 200])
```

```python
labels=['light_weight','medium_weight','heavy_weight']
weights=[light_weight,medium_weight,heavy_weight]
plt.pie(weights,labels=labels)
plt.show()
```

```python
barcelona = fifa.loc[fifa['Club'] == 'FC Barcelona']['Overall'] 
madrid = fifa.loc[fifa['Club'] == 'Real Madrid']['Overall']
juventus = fifa.loc[fifa['Club'] == 'Juventus']['Overall']
labels = ['FC Barcelona', 'Real Madrid', 'Juventus']    
plt.boxplot([barcelona, madrid, juventus], labels = labels)
plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')
plt.show()
```
